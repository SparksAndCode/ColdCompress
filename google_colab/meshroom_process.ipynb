# Google Colab Meshroom 

This is an example I found on the internet explaining how to use Google Colab to process images with a CUDA card and get a point cloud back. If you have a CUDA card in your computer then you don't need all this just download Meshroom and run it locally.

1. You will need to create a .mg project file locally
    * Open meshroom -> import all your images -> save project as .mg file (i called mine face.mg, as you can see below in the command line calls)
    * Once you have the .mg file then you need to open it up and edit it. 
      * Find where the filepaths are defined and modify them to point to "/content/drive/MyDrive/cuda/images/<orig_filename>"
2. Open your google drive, create a folder called "cuda". 
    * Upload your modified Meshroom project file there.
3. Next, create a folder called "images" in the "cuda" folder. 
    * Upload all your images you imported into the meshroom project to that folder.

At this point you should have everything you need to start executing the code blocks below. Keep the tab open and check on it so you don't time out.
It took me a day or so of executing the blocks here and there to complete all the commands below.

Lastly, look in "cuda/MeshroomCache/Texturing/<some_hash>/" and you should see a "texturedMesh.obj". You can take that and start post-processing with meshlab or whatever.
